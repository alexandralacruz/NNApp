1. Implementación de un percetron simple
En esta guía aprenderás a construir una neurona perceptrónclases, utilizando únicamente

El ejemplo práctico es una compuerta lógica AND.

```python
X = np.array([
[0, 0],
[0, 1],
[1, 0],
[1, 1]
])
y_and = np.array([0, 0, 0, 1])
y_or = np.array([0, 1, 1, 1])
y_xor = np.array([0, 1, 1, 0])
```

pesos = np.zeros(2)

1.5. Entrenamiento del perceptrón

Entrenamos el perceptrón durante un número determinado de épocas:

Entrenamiento del perceptrón Entrenamos el perceptrón durante un número determinado de épocas: https://aprendeia.com/que-es-el-perceptron-simple-y-multicapa/

```python
epocas = 10
for epoca in range(epocas):
  for i in range(len(X)):
    # Recorremos cada ejemplo de entrenamiento
    entrada = X[i] # Obtenemos una muestra de entrada (par de bits)
    salida_esperada = y_and[i] # Valor de salida esperado para esa entrada
    # suma ponderada (Figura 2)
    # Producto punto entre entrada y pesos + sesgo
    suma_ponderada = np.dot(entrada, pesos) + bias
    # Activación: función escalón
    salida = step(suma_ponderada) # Aplicamos la función escalón (0 o 1)
    # Cálculo del error
    # Diferencia entre la salida esperada y la predicción
    error = salida_esperada - salida
    # Regla de aprendizaje del perceptrón
    pesos += lr * error * entrada # Actualizamos los pesos
    bias += lr * error # Actualizamos el sesgo (bias)
    # Mostrar los pesos y sesgo actualizados en cada época
    print(f"Época {epoca+1}: Pesos = {pesos}, Bias = {bias}")
```

```python

np.random.seed(0) # pesos y bias en 0
W1 = np.random.randn(2, 4)
b1 = np.zeros((1, 4))
W2 = np.random.randn(4, 1)
b2 = np.zeros((1, 1))
def sigmoid(x):
  return 1 / (1 + np.exp(-x))

def sigmoid_deriv(x):
  return sigmoid(x) * (1 - sigmoid(x))

def relu(x):
    return np.maximum(0, x)

def relu_deriv(x):
    return np.where(x > 0, 1, 0)

def tanh(x):
    return np.tanh(x)

def tanh_deriv(x):
    return 1 - np.tanh(x) ** 2


```

```python
lr = 0.1
epochs = 10000

for epoch in range(epochs):
  # suma ponderada neurona 1 y aplicación de la función sigmoide
  Z1 = np.dot(X, W1) + b1
  A1 = relu(Z1)
  # suma ponderada neurona 2 y aplicación de la función sigmoide
  Z2 = np.dot(A1, W2) + b2
  A2 = sigmoid(Z2)
  # Calcular el error (diferencia entre la salida esperada y la predicción)
  error = y_xor.reshape(-1,1) - A2
  # Calcular la pérdida (mean squared error). Función de costo
  loss = np.mean(error ** 2)
  # Derivada del error respecto a la salida de la red (dA2)
  # Se multiplica por la derivada de la función de activación (sigmoide)
  # para aplicar el descenso del gradiente
  # Retropropagación

  dA2 = error * relu_deriv(Z2)
  # Gradiente de los pesos de la segunda capa (W2)
  dW2 = np.dot(A1.T, dA2)# Gradiente del sesgo de la segunda capa (b2)
  db2 = np.sum(dA2, axis=0, keepdims=True)
  # Retropropagar el error a la capa oculta
  dA1 = np.dot(dA2, W2.T) * tanh_deriv(Z1)
  # Gradiente de los pesos de la primera capa (W1)
  dW1 = np.dot(X.T, dA1)
  # Gradiente del sesgo de la primera capa (b1)
  db1 = np.sum(dA1, axis=0, keepdims=True)
  # Actualizar los pesos y los sesgos con descenso de gradiente
  W2 += lr * dW2
  b2 += lr * db2
  W1 += lr * dW1
  b1 += lr * db1
  # Imprimir pérdida cada 10000 épocas
  if epoch % 10000 == 0:
        print(f"Epoch {epoch}, Loss: {loss:.4f}")
```

```python
print("\nResultados de la red después del entrenamiento:")
for i in range(len(X)):
  z1 = np.dot(X[i], W1) + b1
  a1 = relu(z1)
  z2 = np.dot(a1, W2) + b2
  a2 = sigmoid(z2)
  print(f"Entrada: {X[i]} -> Salida: {a2[0][0]:.4f}")
#Resultado esperado (aproximado)
#Entrada: [0 0] -> Salida: ~0
#Entrada: [0 1] -> Salida: ~1
#Entrada: [1 0] -> Salida: ~1
#Entrada: [1 1] -> Salida: ~0
```
